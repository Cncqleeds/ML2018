```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
cluster1 = np.random.uniform(0.5,1.5,(2,10))
cluster2 = np.random.uniform(5.5,6.5,(2,15))
cluster3 = np.random.uniform(3.,4.,(2,8))
```

```python
x = np.hstack((cluster1,cluster2,cluster3)).T
```

```python
plt.scatter(x[:,0],x[:,1])
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

### 测试9中不同聚类中心下的聚类质量

```python
K = range(1,10)

meandistortions = []
```

```python
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(x)
#     print(kmeans.cluster_centers_)
#     print()

#     print(cdist(x,kmeans.cluster_centers_,'euclidean'))
#     print()
    index = sum(np.min(cdist(x,kmeans.cluster_centers_,'euclidean'),axis=1))/x.shape[0]
    meandistortions.append(index)
    
plt.plot(K,meandistortions,'bx-')
plt.xlabel('k')
plt.ylabel('Average Dispersion')
plt.title('Selecting k with the Elbow Method')
plt.show()
```
