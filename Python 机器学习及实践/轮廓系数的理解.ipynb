```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
x1 = np.array([1,2,3,1,5,6,5,5,3,7,8,9,7,9])
x2 = np.array([1,3,2,2,8,6,7,6,7,1,2,1,1,3])

```

```python
x = np.array(list(zip(x1,x2)))
```

```python
colors = ['b','g','r','c','m','y','k','b']
markers = ['o','s','D','v','^','p','*','+']
```

```python
clusters = [2,3,4,5,8]
```

#### 分割 3 x 2  = 6个子图

```python
plt.subplot(3,2,1)
```

```python
plt.figure(figsize=(16,11))
plt.subplot(3,2,1)


plt.xlim([0,10])
plt.ylim([0,10])

plt.title('Instances')
plt.scatter(x1,x2)

clusters = [2,3,4,5,8]
subplot_counter = 1
sc_scores = []

for t in clusters:
    subplot_counter += 1
    plt.subplot(3,2,subplot_counter)
    kmeans_model = KMeans(n_clusters=t).fit(x)
    
    for i,label in enumerate(kmeans_model.labels_):
        plt.plot(x1[i],x2[i],color=colors[label],marker=markers[label],ls='None')
        plt.xlim([0,10])
        plt.ylim([0,10])
        
    sc_score = silhouette_score(x,kmeans_model.labels_)

    sc_scores.append(sc_score)

    plt.title("K=%s, sil_coef=%0.03f"%(t,sc_score))
plt.figure()
plt.plot(clusters,sc_scores,'*-')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient Score")
plt.show()
plt.show()
```
