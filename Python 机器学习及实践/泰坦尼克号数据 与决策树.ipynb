```python
URL = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic.txt"
titanic = pd.read_csv(URL)
```

```python
titanic.info()
```

#### 数据预处理

数据决定了学习效果的上限，算法能够逼近这个上限

```python
titanic[titanic['age'].isnull()]
```

```python
x = titanic[['pclass','age','sex']]
y = titanic['survived']
```

```python
x.info()
```

#### 一种预处理的方法，填充空值为某个值

```python
x['age'].fillna(x['age'].mean(),inplace=True)
```

```python
x.info()
```

```python
# 数据分割

from sklearn.cross_validation  import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .25,random_state=2018)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
x.info()
```

#### 类别类型的特征转换

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
vec = DictVectorizer(sparse=False)
```

```python
# 注意用法
x_train = vec.fit_transform(x_train.to_dict(orient='record'))
x_test = vec.transform(x_test.to_dict(orient='record'))
```

```python
type(x_train)
```

```python
vec.feature_names_
```

#### 使用决策树模型进行预测

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()

dtc.fit(x_train,y_train)
```

```python
y_test_pred = dtc.predict(x_test)
```

```python
dtc.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test_pred, y_test,target_names=['died','survivied']))
```
