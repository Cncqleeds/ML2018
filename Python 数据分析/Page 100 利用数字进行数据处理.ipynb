```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
points = np.arange(-5,5,0.01)

xs,ys = np.meshgrid(points,points)

z = np.sqrt(9*xs ** 2 + 4 *ys ** 2)

plt.imshow(z, cmap = plt.cm.coolwarm); plt.colorbar()
plt.title(" Image plot of $\sqrt{9x^2 + 4y^2}$ for a grid of values")
plt.grid()
plt.show()
```

```python
position = 0
walk = [position]
steps = 100

for i in range(steps):
    step = 1 if np.random.randint(0,2,1)[0] else -1
    position += step
    walk.append(position)
```
